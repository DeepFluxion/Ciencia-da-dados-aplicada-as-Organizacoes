## Matplotlib

O matplotlib é uma biblioteca com recursos para a geração de gráficos 2D a partir de arrays. Gráficos comuns podem ser criados com alta qualidade a partir de comandos simples, inspirados nos comandos gráficos do MATLAB.

O matplotlib pode ser conceitualmente dividido em 3 partes:

pylab: conjunto de funções disponíveis em matplotlib.pylab que permite a geração de código similar ao MATLAB.

frontend ou API: conjunto de classes que realizam o trabalho pesado, criando as figuras, texto, linhas etc. Essa é uma interface abstrata que independe da saída.

backends: conjunto de funções que dependem do dispositivo de saída (display). Por exemplo, PS para gráficos em PostScript, SVG gera gráficos em Scalable Vector Graphics, Agg cria figuras no formato PNG, GTK permite que os gráficos sejam incluídos em aplicações GTK+, e assim para PDF, WxWidgets, Tkinter, etc.

## Arrays NumPy
A estrutura de dados mais importante que o NumPy fornece é um objeto poderoso, um tipo de Array, chamada ndarray. 

O objeto ndarray consiste em um segmento unidimensional contíguo da memória do computador, combinado com um esquema de indexação que mapeia cada item para um local no bloco de memória. 

O bloco de memória contém os elementos em uma ordem principal de linha (estilo a linguagem C) ou uma ordem de coluna principal (estilo FORTRAN ou MATLAB), mas não vamos nos aprofundar tanto nesses detalhes. 

Essa Array NumPy é uma tabela de elementos (geralmente números), todos do mesmo tipo, indexados por uma tupla de inteiros positivos. No NumPy, as dimensões são chamadas de eixos. 

No exemplo mostrado abaixo, a Array tem 2 eixos. O primeiro eixo tem um comprimento de 2, o segundo eixo tem um comprimento de 3:

Outra facilidade proporcionada pela estrutura são seus métodos que fornecem informações estatísticas sobre os valores, como média .mean() e desvio padrão .std(). Encorajo o leitor(a) a investigar e verificar alguns dos métodos e atributos da estrutura usando o TAB para auto-completação na shell do Python, ou simplesmente checar a completíssima documentação oficial deste objeto.

## Noçoes de Matplotlib

Vamos começar com alguns dados  que vamos usar para fazer alguns exemplos

```python
#definindo variáveis que vamos usar nos exemplos
x = [1, 2, 3]
y = [10, 20, 30]
y2 = [15, 10, 40]
y3 = [20, 10, 35]
yBar = [3, 10, 7, 5, 3, 4.5, 6, 8.1]
z = [i * 1.5 for i in yBar]
xBar = range(len(yBar))
azul = "blue"
verde = "green"
preto = "black"
```

## Desenhando uma linha

```python
plt.plot(x, y)
```

### Adicionado rótulos

```python
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Título do gráfico')
plt.plot(x, y)
```

### Grafico com mais de uma linha

```python
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gráfico com mais de uma linha')
plt.plot(x, y)
plt.plot(x, y2)
plt.plot(x, y3)
```

### Grafico de Barras

```python
plt.bar(xBar, yBar, color=azul)
plt.title('Gráfico de Barras')
plt.xlabel('Xbar')
plt.ylabel('Ybar')
```

## Gráfico de Barras e linha

```python
plt.bar(xBar, yBar, color='gray', align='center')
plt.plot(xBar, z, color='yellow', marker='o', linestyle='--', linewidth=3)
plt.title('Gráfico de Barras e linha')
plt.xlabel('Xbar')
plt.ylabel('Ybar')
```

# Parte 2 Trabalhando com Arvores de Decisão

## Para fazer a modelagem nós vamos sempre seguir um fluxo que consiste nas seguintes etapas

1. Criação dos dados X e y ( y é o atributo alvo e X são os atributos usado no modelo)
2. Separação dos dados em grupo de treinamento e grupo de teste  ou seja o X e y serão separados em X_train, y_train e X_test, y_test.
2. Criação do modelo (usando uma classe do scikit learn nós vamos criar o modelo com com as características que queremos)
3. Treinamento do modelo fazendo um **model.fit(X_train,y_train)**
4. Teste do modelo **model.fit(X_train,y_train).predict(X_test,y_test)**
5. Visualização e análise

## 1. Criação dos dados X e y ( y é o atributo alvo e X são os atributos usado no modelo)

Os dados podem vir de diversas fontes, como dataFrames do Pandas, planilhas do excele, arquivos csv, entre outros...
O mais importante é verificar qual será o **atributo alvo** e os **atributos preditores** que serão usado no modelo com preditores. O **atributo alvo** é sempre único e consiste naquilo que desejamos predizer. Os **atributos preditores** são aqueles que usraemos para prever os novos valores do atributo alvo qunado este estiver em produção. Vamos como isso funciona usando um conjunto de dados disponível no próprio Scikit Learn, que é o data set iris

```python
from sklearn.datasets import make_moons, make_circles, make_classification
from sklearn.model_selection import train_test_split
from sklearn import tree
```

## Experimento 1

## Criação dos dados

```python
X, y = make_classification(n_features=2, n_redundant=0, n_informative=2,
                           random_state=1, n_clusters_per_class=1, n_samples=1000)
rng = np.random.RandomState(2)
X += 2 * rng.uniform(size=X.shape)
```

## Separação dos dado em um Grupo de Treinamento(60%) e um de Teste (40%)

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.4, random_state=42)
```

## Visualização dos Dados

```python
plt.figure(figsize=(7,7))
plt.scatter(X[:,0],X[:,1], c=y, cmap='viridis')
plt.title('Visualização do Dataset 1')
plt.grid()
plt.xlabel('X1')
plt.ylabel('X2')
```

## Ajuste do Modelo com os dados de treianemnto (X_train, y_train)

```python
model = tree.DecisionTreeClassifier(max_depth=3)
model.fit(X_train, y_train)
score = model.score(X_test, y_test)
```

### Vamos visualizar os dados de teste (y_test)

```python
plt.plot(y_test)
```

### Vamos visualizar os dados previstos model.predict(X_test)



```python
plt.plot(model.predict(X_test))
```

### Vamos tentar melhorar essa visualização vendo os dados (real e previsto) dados em conunto

```python
plt.scatter(y_test, model.predict(X_test))
plt.xlabel('y_test')
plt.ylabel('model.predict(X_test)')
```

### Vamos visualizar a diferença entre os dados de teste (y_test) e os dados ajustados para X_test **(model.predict(X_test)**

```python
plt.plot((y_test-model.predict(X_test)))
```

Os dados tanto de teste que são os dados reais e os dados previstos com o modelo treinado podem ser 0 ou 1. Portanto temos três situações:
1 - O dado real (y_test) é igual ao dado previsto e adiferença é zero, ou seja o modelo acertou.
2 - O dado real (y_test) é diferente ao dado previsto e a diferença pode ser 1 ou -1, ou seja o modelo errou.

Vamos visualizar a soma dos erros ou seja fazer uma soma progressiva de modo que se o modelo acertou temos 0, se errou temo -1 ou 1

```python
plt.plot((y_test-model.predict(X_test).cumsum()))
```

Ou seja é muito dificil de visulaizar o erro portanto temos o valor de score que diz qual é a procentagem de acertos da classificação

##Visualização da distribuição da classsficação esta é uma função desenolvida por Jake Van der Plas para visualizar os limtes de classficação

```python
def visualizar_modelo(model, X, y, cm='viridis'):
    h = .02
    x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
    y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
    plt.figure(figsize=(7,7))
    plt.contourf(xx,yy,Z, alpha =0.5)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cm, edgecolors='k')

    plt.show()
```

```python
visualizar_modelo(model, X, y, cm='viridis')
```

## Agora vamos visualizar a árvore de classficação

```python
plt.figure(figsize=(20,7))
tree.plot_tree(model.fit(X,y), fontsize=10);
```
